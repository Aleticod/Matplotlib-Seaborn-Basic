# Grafico de barras
Para este tipo de graficas se necesitan variables de tipo categoricas y numericas

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Variables categoricas

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 1200, 400, 1400]
```

```python
plt.bar(country, population, width=0.5, color=['red', 'blue', '#aa44aa', 'cyan', 'magenta'])
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.title('Country Population')
plt.xlabel('Countries')
plt.ylabel('Population')
plt.show()
```

```python
plt.barh(country, population)
plt.show()
```
