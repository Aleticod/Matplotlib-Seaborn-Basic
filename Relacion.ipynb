# Tipo de grafica relacion

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns. scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.48))
plt.show()
```

```python
plt.figure(figsize=(8,8))
markers = {'Lunch':'D', 'Dinner':'s'}
sns. scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12,0.48))
plt.show()
```

```python
sns.lineplot(data=tips, x = 'total_bill', y = 'tip')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', style='time', col='time')
plt.show()
```

```python
sns.pairplot(data=tips, hue='day')
plt.show()
```
