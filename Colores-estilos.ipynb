# Colores y estilos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 20)
```

```python
print(plt.style.available)
```

```python
plt.style.use('ggplot')
```

## Modificaciones de tipo Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, x+ 1, 'r--')
ax.plot(x, x+ 2, 'b-')
ax.plot(x, x+ 3, 'y.-')
ax.plot(x, x+ 4, 'c:')
ax.plot(x, x+ 5, 'mD')
```

## Modificaciones con Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, x+ 1, color='green', alpha=0.1, linewidth=8, linestyle='dashed')
ax.plot(x, x+ 2, color = 'cyan', marker='o', markersize=10, markerfacecolor='#aaffcc')
ax.plot(x, x+ 3, color='#66ff66')
ax.plot(x, x+ 4, color='#ffccaa')
ax.plot(x, x+ 5, color='#000000')
```
