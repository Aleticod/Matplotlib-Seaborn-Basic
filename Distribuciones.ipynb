# Distribuciones

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', alpha=0.5, stat='count', multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True, bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='kde', multiple='stack')
plt.show()
```
